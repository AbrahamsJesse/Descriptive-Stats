## Challenge 1
#### 1.- Define a function that simulates rolling a dice 10 times. Save the information in a dataframe.
**Hint**: you can use the *choices* function from module *random* to help you with the simulation.

```python
# your code here
import matplotlib.pyplot as plt
import pandas as pd
import random

lista = [1,2,3,4,5,6]
def new_function (number):
    lista2=[]

    for item in range (10):
        great= random.choices (number, weights=None, cum_weights=None, k=1)
        lista2.append(great[0])
    return lista2



lista3=new_function(lista)

print(lista3)
```

#### 2.- Plot the results sorted by value.

```python

#Plotting
import pandas as pd
df = pd.DataFrame (lista3, columns=['Dice results'])
df.sort_values(by = ['Dice results'], inplace=True)
df.plot(kind='bar')
```

#### 3.- Calculate the frequency distribution and plot it. What is the relation between this plot and the plot above? Describe it with words.

```python

#Getting the frequency distribution
import pandas as pd
frequency_distribution = pd.DataFrame({'Dice results': lista3}).value_counts().reset_index(name='Count')
frequency_distribution 
```

```python

#Plotting
import seaborn as sns

sns.distplot(frequency_distribution )
plt.show()
```
